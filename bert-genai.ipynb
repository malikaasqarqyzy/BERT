```python
import pandas as pd

# Read the file with more lenient parsing
df = pd.read_csv('/kaggle/input/dataset1/dataset (1).csv', 
                 engine='python',  
                 encoding='utf-8',
                 on_bad_lines='skip',    # For newer pandas versions
                 quoting=3)  # Minimal quoting

# Print out details
print(f"Number of rows: {len(df)}")

# Count the total lines in the original file
with open('/kaggle/input/dataset1/dataset (1).csv', 'r', encoding='utf-8') as file:
    total_lines = sum(1 for line in file)

print(f"Total lines in original file: {total_lines}")

# Display sentiment distribution
print("\nSentiment Distribution:")
print(df['Sentiment'].value_counts())
print("\nSentiment Percentage:")
print(df['Sentiment'].value_counts(normalize=True) * 100)

# Optional: Investigate any potential issues
def check_file_integrity(filepath):
    print("\nFile Integrity Check:")
    with open(filepath, 'r', encoding='utf-8') as file:
        for i, line in enumerate(file, 1):
            try:
                # Try to split the line
                parts = line.strip().split(',')
                if len(parts) < 2:
                    print(f"Problematic line {i}: {line.strip()}")
            except Exception as e:
                print(f"Error processing line {i}: {e}")

check_file_integrity('/kaggle/input/dataset1/dataset (1).csv')
```
